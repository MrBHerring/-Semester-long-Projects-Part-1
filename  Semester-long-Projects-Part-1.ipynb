# Creating Graphical Representations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
filename = 'C:\\Users\\frost\\Desktop\\Fall 2023\\CSC 510\\project\\cleaned-police-violence-shootings-2015-2019.csv'
data = pd.read_csv(filename)

#Display number of Fatalities by Race
plt.figure(figsize=(10, 10)) 
sns.countplot(data=data, x='race') #Using Seaborn, produce a graph to visualize fatalities by Race
plt.title('Number of Fatalities by Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()


#Display number of Fatalities by Gender
plt.figure(figsize=(10, 10))
sns.countplot(data=data, x='gender') #Similar results, Gender becomes the next variable
plt.title('Number of Fatalities by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

#Display number of Fatalities by Year
plt.figure(figsize=(10, 10))
sns.countplot(data=data, x='year') #Year becomes the last variable 
plt.title('Number of Fatalities by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

# Presenting police shootings by race and year within each category.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
filename = 'C:\\Users\\frost\\Desktop\\Fall 2023\\CSC 510\\project\\cleaned-police-violence-shootings-2015-2019.csv'
data = pd.read_csv(filename)

# Define a list of unique racial categories in the dataset
unique_races = data['race'].unique()

#Create a graph for each race
for race in unique_races:
    # Filter the data for the specific race
    race_data = data[data['race'] == race]
    
    #Display the distribution of shootings by year for the specific race
    plt.figure(figsize=(8, 4))
    sns.countplot(data=race_data, x='year')
    plt.title(f'Distribution of Police Shootings for {race} by Year')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.show()
```

## Using the unique function and for loop, a graph for each race category can be produced.

# Addressing two important questions

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Load the data
filename = 'C:\\Users\\frost\\Desktop\\Fall 2023\\CSC 510\\project\\cleaned-police-violence-shootings-2015-2019.csv'
data = pd.read_csv(filename)

#Display statistics on police shootings and mental health issues
plt.figure(figsize=(10, 6))
mental_illness_by_race = data[data['signs_of_mental_illness'] == True]
sns.countplot(data=mental_illness_by_race, x='race')
plt.title('Number of Fatalities with Mental Health Issues by Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()



# Place all weapons in one separated category called "armed"
def categorize_armed_status(armed_status):
    if armed_status == 'unarmed':
        return 'unarmed'
    else:
        return 'armed'

data['armed'] = data['armed'].apply(categorize_armed_status)

#Display statistics on armed or unarmed victims
plt.figure(figsize=(10, 10))
sns.countplot(data=data, x='armed', hue='manner_of_death')
plt.title('Outcomes of Police Shootings by Armed Status (Combined)')
plt.xlabel('Armed Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Manner of Death')
plt.show()
```

# Question 1: Are whites or blacks more likely to killed by a police officer due to the victim's mental health?

## Looking at the graph concerning mental health issues, The data shows that whites are more prone to getting shot than any other race.
